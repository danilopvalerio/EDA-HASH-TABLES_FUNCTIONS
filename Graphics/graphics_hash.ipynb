```python
import matplotlib.pyplot as plt
import numpy as np

def plot_collisions():
    techniques = ['Sond.Linear', 'Sond.Quadrática', 'Sond.Dupla']
    collisions = [72931, 57131, 51665]

    x = np.arange(len(techniques))
    width = 0.6

    fig, ax = plt.subplots(figsize=(12, 8))  # Aumentando o tamanho da figura
    rects1 = ax.bar(x, collisions, width, color='orange')

    ax.set_xlabel('Técnica de Hashing')
    ax.set_ylabel('Quantidade de Colisões')
    ax.set_title('Colisões por Técnica de Hashing')
    ax.set_xticks(x)
    ax.set_xticklabels(techniques, rotation=45, ha='right')

    for rect in rects1:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

    fig.tight_layout()
    plt.show()

# Exibir o gráfico
plot_collisions()
```

```python
import matplotlib.pyplot as plt

# Dados dos algoritmos
algorithm_names = ['Endereço fechado', 'Sond. Linear', 'Sond. Quadrática', 'Sond. Dupla']
insertion_times = [25, 400, 26, 19]
search_times = [33, 37, 32, 16]

# Configuração dos dados
x = range(len(algorithm_names))
bar_width = 0.35

# Criando a figura e os eixos para os dois subplots
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(10, 12))

# Plotando gráfico para os tempos de inserção
bars1 = ax1.bar(x, insertion_times, bar_width, label='Tempos de Inserção', color='b')

# Adicionando detalhes ao primeiro gráfico
ax1.set_xlabel('Algoritmo')
ax1.set_ylabel('Tempo (ms)')
ax1.set_title('Tempos de Inserção por Algoritmo')
ax1.set_xticks(x)
ax1.set_xticklabels(algorithm_names)
ax1.legend()

# Plotando gráfico para os tempos de busca
bars2 = ax2.bar(x, search_times, bar_width, label='Tempos de Busca', color='g')

# Adicionando detalhes ao segundo gráfico
ax2.set_xlabel('Algoritmo')
ax2.set_ylabel('Tempo (ms)')
ax2.set_title('Tempos de Busca por Algoritmo')
ax2.set_xticks(x)
ax2.set_xticklabels(algorithm_names)
ax2.legend()

# Ajustes de layout para melhor visualização
plt.xticks(rotation=45)
plt.tight_layout()

# Exibindo os gráficos
plt.show()
```

```python
import openpyxl

# Criar um novo arquivo Excel
wb = openpyxl.Workbook()
sheet = wb.active
sheet.title = "Hashing Comparativo"

# Dados para inserção nas tabelas
dados_encadeamento_fechado = {
    "Tipo de Operação": ["Tempo de Inserção", "Tempo de Busca"],
    "Tempo": [25, 33]
}

dados_sondagem_linear = {
    "Tipo de Operação": ["Tempo de Inserção", "Tempo de Busca", "Colisões", "Rehashes"],
    "Tempo": [400, 37, 72931, 4]
}

dados_sondagem_quadratica = {
    "Tipo de Operação": ["Tempo de Inserção", "Tempo de Busca", "Colisões", "Rehashes"],
    "Tempo": [26, 32, 57131, 4]
}

dados_sondagem_dupla = {
    "Tipo de Operação": ["Tempo de Inserção", "Tempo de Busca", "Colisões", "Rehashes"],
    "Tempo": [19, 16, 51665, 4]
}

# Escrever os dados no arquivo Excel
def escrever_dados(sheet, dados, row_start, col_start):
    for i, tipo_operacao in enumerate(dados["Tipo de Operação"]):
        sheet.cell(row=row_start + i, column=col_start).value = tipo_operacao
        sheet.cell(row=row_start + i, column=col_start + 1).value = dados["Tempo"][i]

# Escrever os dados para cada tipo de hashing
escrever_dados(sheet, dados_encadeamento_fechado, 1, 1)
escrever_dados(sheet, dados_sondagem_linear, 6, 1)
escrever_dados(sheet, dados_sondagem_quadratica, 11, 1)
escrever_dados(sheet, dados_sondagem_dupla, 16, 1)

# Salvar o arquivo Excel
wb.save("dados_hashing.xlsx")
print("Arquivo Excel 'dados_hashing.xlsx' criado com sucesso.")
```
